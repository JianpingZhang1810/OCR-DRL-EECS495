```python
data_path = "/clever/northwestern/jianping/CelebA/dataset"
data_path_list = [data_path+"/ICDAR2013/ICDAR2013_train/background",
                  data_path+"/ICDAR2013/ICDAR2013_test/background",
                  data_path+"/ICDAR2015/train/background",
                  data_path+"/ICDAR2015/test/background",
                  data_path+"/SVT/background"]
print(data_path_list)
```

```python
import os
whole_data_path = []
for path in data_path_list:
    for fpathe,dirs,fs in os.walk(path):
        for f in fs:
            #print(f)
            if f[-4:] == ".jpg":
                print(os.path.join(fpathe,f))
                whole_data_path.append(os.path.join(fpathe,f))
            
```

```python
print(len(whole_data_path))
```

```python
import random
random.shuffle(whole_data_path)
```

```python
split = int(len(whole_data_path)*0.8)
```

```python
with open('split.txt', 'w') as f:
    for i in range(len(whole_data_path)):
        f.write(whole_data_path[i])
        f.write(" ")
        if i <= split:
            f.write("0")
        else:
            f.write("2")
        f.write("\n")
```
