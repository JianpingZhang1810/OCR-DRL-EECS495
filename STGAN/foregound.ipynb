```python
import numpy as np
import scipy.misc

#celebA_path = "/clever/northwestern/jianping/CelebA/dataset"
output_path = "dataset"
data_path = "/clever/northwestern/jianping/CelebA/dataset"
data_path_list = [data_path+"/ICDAR2013/ICDAR2013_train/foreground",
                  data_path+"/ICDAR2013/ICDAR2013_test/foreground",
                  data_path+"/ICDAR2015/train/foreground",
                  data_path+"/ICDAR2015/test/foreground",
                  data_path+"/SVT/foreground"]
print(data_path_list)
```

```python
import os
whole_data_path = []
for path in data_path_list:
    for fpathe,dirs,fs in os.walk(path):
        for f in fs:
            #print(f)
            if f[-4:] == ".jpg":
                print(os.path.join(fpathe,f))
                whole_data_path.append(os.path.join(fpathe,f))
```

```python
print(len(whole_data_path))
```

```python
import random
random.shuffle(whole_data_path)

selected = whole_data_path[:50]
```

```python
img = scipy.misc.imread(selected[0])
final_img = np.zeros((len(selected),img.shape[0],img.shape[1],4),dtype="uint8")
for batch in range(len(selected)):
    img = scipy.misc.imread(selected[batch])
    mask = np.zeros((img.shape[0],img.shape[1],1),dtype="uint8")
    for i in range(img.shape[0]):
        for j in range(img.shape[1]):
            if img[i,j,0] != 255 and img[i,j,1] != 255 and img[i,j,2] != 255:
                mask[i][j] = 255
    final = np.concatenate((img,mask),axis = 2)
    final_img[batch,:,:,:] = final 
```

```python
print(final_img.shape)
```

```python
np.save("{0}/foreground.npy".format(output_path),final_img)
```
